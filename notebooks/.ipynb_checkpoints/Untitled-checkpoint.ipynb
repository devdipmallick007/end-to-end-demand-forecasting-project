```python
#use only useable column

df_orders = df_orders[['order_id','customer_id','order_date','store_id','order_total']]
df_customers = df_customers[['customer_id','area','customer_segment','total_orders','avg_order_value']]
df_products = df_products[['product_id','category','brand','price','mrp','shelf_life_days','margin_percentage']]
df_weather = df_weather[['area','date','temperature','precipitation']]
df_order_items = df_order_items[['order_id','product_id','quantity','unit_price']]
df_marketing_performance = df_marketing_performance[['date','spend','impressions','conversions','revenue_generated']]
df_inventory = df_inventory[['product_id','date','stock_received','damaged_stock']]
```

```python
#bottom-level series

# Filter delivered orders only
df_delivered = df_orders.copy()  # assuming you've already filtered usable columns

# Merge with order items
merged = df_order_items.merge(
    df_delivered[['order_id','order_date','store_id']],
    on='order_id', how='inner'
)

# Convert order_date to datetime
merged['date'] = pd.to_datetime(merged['order_date']).dt.date

# Compute revenue
merged['revenue'] = merged['quantity'] * merged['unit_price']

# Aggregate to bottom-level series
product_store_day = (
    merged.groupby(['date','store_id','product_id'], as_index=False)
    .agg(qty=('quantity','sum'), revenue=('revenue','sum'))
)

print("Bottom-level series shape:", product_store_day.shape)
product_store_day.head()
```

```python
# IQR method
Q1 = product_store_day['qty'].quantile(0.25)
Q3 = product_store_day['qty'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

# Flag and cap outliers
product_store_day['outlier_qty'] = ((product_store_day['qty'] < lower_bound) | (product_store_day['qty'] > upper_bound))
product_store_day['qty'] = np.where(product_store_day['qty'] > upper_bound, upper_bound,
                                   np.where(product_store_day['qty'] < lower_bound, lower_bound,
                                            product_store_day['qty']))
print(product_store_day)
print(product_store_day)
```

```python
#Join product attributes
product_store_day = product_store_day.merge(
    df_products,
    on='product_id', how='left'
)

print(product_store_day)
```

```python
#Join inventory
df_inventory['date'] = pd.to_datetime(df_inventory['date']).dt.date
product_store_day = product_store_day.merge(
    df_inventory,
    on=['product_id','date'], how='left'
)
print(product_store_day)
```

```python
#Join marketing data
df_marketing_performance['date'] = pd.to_datetime(df_marketing_performance['date']).dt.date
daily_marketing = df_marketing_performance.groupby('date')[['spend','impressions','conversions','revenue_generated']].sum().reset_index()
product_store_day = product_store_day.merge(daily_marketing, on='date', how='left')

print(df_marketing_performance)
print(daily_marketing)
print(product_store_day)
```

```python
#Time-series features
#Lags & rolling windows
product_store_day = product_store_day.sort_values(['product_id','store_id','date'])

# Lags
product_store_day['lag_1'] = product_store_day.groupby(['store_id','product_id'])['qty'].shift(1)
product_store_day['lag_7'] = product_store_day.groupby(['store_id','product_id'])['qty'].shift(7)

# Rolling sums
product_store_day['rolling_7_sum'] = product_store_day.groupby(['store_id','product_id'])['qty'].transform(lambda x: x.shift(1).rolling(7).sum())
product_store_day['rolling_28_sum'] = product_store_day.groupby(['store_id','product_id'])['qty'].transform(lambda x: x.shift(1).rolling(28).sum())

```

```python
#Calendar features
product_store_day['date'] = pd.to_datetime(product_store_day['date'])
product_store_day['weekday'] = product_store_day['date'].dt.weekday
product_store_day['is_weekend'] = product_store_day['weekday'].isin([5,6]).astype(int)
product_store_day['month'] = product_store_day['date'].dt.month
```

```python
store_area_map = df_customers.groupby('customer_id')['area'].first().to_dict()
# If store_id maps differently, adjust
product_store_day['area'] = product_store_day['store_id'].map(store_area_map)
df_weather['date'] = pd.to_datetime(df_weather['date']).dt.date
product_store_day = product_store_day.merge(
    df_weather[['area','date','temperature','precipitation']],
    on=['area','date'], how='left'
)
```
